```python
import requests

def get_weather(city):
   url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=79007029562c2ab46617d801d8018750&units=metric"

   try:
       response= requests.get(url)
       weather_data = response.json()

       temperature = weather_data["main"]["temp"]
       humidity = weather_data["main"]["humidity"]
       max_temp = weather_data["main"]["temp_max"]
       min_temp = weather_data["main"]["temp_min"]
       wind_speed = weather_data["wind"]["speed"]
       pressure = weather_data["main"]["pressure"]


       print(f"Weather in {city}:")
       print(f"Temperature {temperature}°C:")
       print(f"Humidity : {humidity}%")
       print(f"Max Temperature: {max_temp}°C:")
       print(f"Min Temperature: {min_temp}°C:")
       print(f"Wind Speed: {wind_speed} m/s:")
       print(f"pressure : {pressure}")

   except requests.exceptions.RequestException:
        print("Error occured")


#usage
city_name = input("Enter the name of city:")
get_weather(city_name)
```

```python
import gradio as gr
import requests
import os

def get_weather(city):
    api_key = os.getenv("OPENWEATHER_API_KEY")
    if not api_key:
        return "Error: API key not found. Set OPENWEATHER_API_KEY as an environment variable."

    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    data = response.json()

    if data.get("cod") == 200:
        temp = data["main"]["temp"]
        description = data["weather"][0]["description"]
        return f"The current temperature in {city} is {temp}°C with {description}."
    else:
        return f"Sorry, I couldn't find weather information for '{city}'."

# ------------------------
#   Professional UI
# ------------------------

with gr.Blocks(css="""
#title {
    text-align: center;
    font-size: 36px;
    font-weight: bold;
    margin-top: 20px;
}
#subtitle {
    text-align: center;
    font-size: 18px;
    color: #555;
    margin-bottom: 20px;
}
#logo {
    display: block;
    margin-left: auto;
    margin-right: auto;
    width: 120px;
}
""") as demo:

    # --- Logo ---
    gr.Image(
        value="https://i.imgur.com/8Y2YqXK.png",   # A clean sun-rise icon (replace if needed)
        elem_id="logo",
        show_label=False
    )

    # --- Title ---
    gr.HTML("<div id='title'>Naser Weather Bot</div>")
    gr.HTML("<div id='subtitle'>Simple • Fast • Accurate Weather Updates</div>")

    with gr.Row():
        city_input = gr.Textbox(label="Enter City Name")
        get_btn = gr.Button("Get Weather")

    output = gr.Textbox(label="Weather Result", lines=3)

    get_btn.click(fn=get_weather, inputs=city_input, outputs=output)

demo.launch()
```
